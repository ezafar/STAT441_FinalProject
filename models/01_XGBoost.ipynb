```python
import os
import pandas as pd
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
```

```python
X_train = pd.read_csv('../data/X_train.csv')
X_test = pd.read_csv('../data/X_test.csv')
y_train = pd.read_csv('../data/y_train.csv')
X_columns_to_drop = ['id']
y_columns_to_drop = ['id']
X_test.drop(X_columns_to_drop, inplace=True, axis=1)
y_train.drop(y_columns_to_drop, inplace=True, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
X_train['c_abrv'] = label_encoder.fit_transform(X_train['c_abrv'])
```

```python
X_train['c_abrv'].isnull().values.any()
X_train['c_abrv'] = X_train['c_abrv'].astype(str)
```

```python
# Set enable_categorical=True when creating the DMatrix
dtrain = xgb.DMatrix(data=X_train, label=y_train, enable_categorical=True)
```

```python
model_xgb = xgb.XGBClassifier()
y_train[y_train==-1] = 0
model_xgb.fit(X_train, y_train)
```
